```python
df.columns[df.isnull().any()]
df.columns=df.columns.str.lower()
df.columns
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
# 1. Strip spaces, remove '$', and handle potential text issues
df['unit_price'] = df['unit_price'].astype(str).str.replace('$', '', regex=False).str.strip()

# 2. Convert to float (errors='coerce' turns any unconvertible text or blanks into NaN safely)
df['unit_price'] = pd.to_numeric(df['unit_price'], errors='coerce')
```

```python
df.info()
```

```python
df['total']=df['unit_price']*df['quantity']
df.head()
```

```python
#to make connection of pythoin ans mysql

from sqlalchemy import create_engine

# 1. Connect to the database we just created
engine_mysql = create_engine("mysql+pymysql://root:123456@localhost:3306/Walmart_db")

# 2. Push your DataFrame ('df') into MySQL as a table named 'walmart'
try:
    df.to_sql(name='walmart', con=engine_mysql, if_exists='replace', index=False)
    print("Success! Data has been pushed to MySQL Workbench.")
except Exception as e:
    print("Error pushing data:")
    print(e)
```

```python
#to export data automatically to mysql
df.to_sql(name='walmart',con=engine_mysql,if_exists='append',index=False)
```
